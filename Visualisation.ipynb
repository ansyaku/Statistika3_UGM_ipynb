```r
# Load necessary library
library(ggplot2)

# Define parameters
df_values <- c(1, 5, 10, 20)  # Different degrees of freedom
f_values <- seq(0.1, 5, length.out = 100)  # Range of F-values

# Create an empty data frame
data_list <- list()

# Generate density values for each degree of freedom
for (df in df_values) {
  density_values <- df(f_values, df1 = df, df2 = df * 2)  # Compute density
  df_data <- data.frame(df = df, f = f_values, density = density_values)
  data_list[[as.character(df)]] <- df_data  # Store in list
}

# Convert list to a single data frame
df_example <- do.call(rbind, data_list)

# Print first few rows
head(df_example)
```

# **theme_minimal()**

```r
# Basic regression line without confidence interval
ggplot(mtcars, aes(x = wt, y = mpg)) + 
  geom_point() + 
  geom_smooth(method = "lm", se = FALSE) + 
  theme_minimal() + 
  labs(title = "Regression Line of MPG vs Car Weight", 
       subtitle = "Without confidence interval",
       x = "Weight (1000 lbs)", 
       y = "Miles per Gallon")
```

* For formal papers: `theme_bw()` or `theme_light()`
* For a clean look: `theme_classic()` or `theme_minimal()`
* For presentations: `theme_dark()`
* For creative plots: `theme_void()`

# **theme_bw()**

```r
# Basic regression line without confidence interval
ggplot(mtcars, aes(x = wt, y = mpg)) + 
  geom_point() + 
  geom_smooth(method = "lm", se = FALSE) + 
  theme_bw() + 
  theme(
      plot.title = element_text(size = 18, face = "bold", hjust = 0.5),
      plot.subtitle = element_text(size = 18, hjust = 0.5, color = "gray50"),
      axis.text = element_text(size = 14),  # Axis number size
      axis.title = element_text(size = 16, face = "bold")) + 
  labs(title = "Regression Line of MPG vs Car Weight", 
       subtitle = "Without confidence interval",
       x = "Weight (1000 lbs)", 
       y = "Miles per Gallon")
```
