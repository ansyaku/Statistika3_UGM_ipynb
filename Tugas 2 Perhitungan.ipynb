Suppose $X$ is a gamma random variable with $\alpha = 2$ and $\beta = 1$.

* (a) Find $P(X < 3)$ by hands (manually, not using computer).
* (b) Use software to plot the pdf with area where $P(X < 3)$ be shaded in blue colour.

$$
\Huge
P(X<3) = F_X(3) = \frac{1}{\theta^2 \Gamma(\alpha)} 3^{\alpha-1} e^{-\frac{3}{\theta}}
$$

$$
P(X<3) = F_X(3) = \frac{1}{\theta^2 \Gamma(\alpha)} 3^{\alpha-1} e^{-\frac{3}{\theta}}
$$

```r
#Define the parameter
alpha = 2
theta = 15 / 60
x = 3
pgamma(q = x, shape = alpha, scale = theta)
```

```r
#######################################################################################
## PLOTTING GAMMA CUMMULATIVE DISTRIBUTION FUNCTION (CDF) 
#######################################################################################
mean(rgamma(n = 10000, shape = alpha, scale = theta) <= x)
library(dplyr)
library(ggplot2)

options(repr.plot.width=20, repr.plot.height=8) # this is just to make the plots wider (extend to the end of laptop display) 
data.frame(x = 0:1000 / 100, prob = pgamma(q = 0:1000 / 100, shape = alpha, scale = theta, lower.tail = TRUE)) %>%
  mutate(Interval = ifelse(x >= 0 & x <= 3, "0 to 3", "other")) %>%
ggplot(aes(x = x, y = prob, fill = Interval)) +
  geom_area(alpha = 0.3) +
  labs(title = "X ~ Gam(alpha = 2, theta = .25)",
       subtitle = "Probability of 10 events in X hours when the mean time to an event is .25 hours.",
       x = "Interval (x)",
       y = "Cum Probability") 
```
